# Importing Pacakges

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cluster import KMeans
```

# Data Pre-Processing

```python
dataset = pd.read_csv("/kaggle/input/mall-customers-csv/Mall_Customers.csv")
x = dataset.iloc[:,[3,4]].values
```

# Elbow Method Graph

```python
wcss = []
for i in range( 1 , 11 ):
    k = KMeans( n_clusters = i , init = 'k-means++' , random_state = 42 , n_init = 10 )
    k.fit(x)
    wcss.append(k.inertia_)

plt.plot( range( 1 , 11 ), wcss )
plt.title(" The Elbow Method Graph ")
plt.xlabel(" No. of Clusters(k) ")
plt.ylabel(" wcss ")
plt.show()
```

# Model Training

```python
k = KMeans( n_clusters = 5 , init = 'k-means++' , random_state = 50 , n_init = 10 )
y_predict = k.fit_predict(x)
```

```python
plt.scatter(x[y_predict == 0 , 0] , x[y_predict == 0 , 1] , s = 100 , c = 'blue' , label = 'Cluster 1')
plt.scatter(x[y_predict == 1 , 0] , x[y_predict == 1 , 1] , s = 100 , c = 'red' , label = 'Cluster 2')
plt.scatter(x[y_predict == 2 , 0] , x[y_predict == 2 , 1] , s = 100 , c = 'green' , label = 'Cluster 3')
plt.scatter(x[y_predict == 3 , 0] , x[y_predict == 3 , 1] , s = 100 , c = 'pink' , label = 'Cluster 4')
plt.scatter(x[y_predict == 4 , 0] , x[y_predict == 4 , 1] , s = 100 , c = 'yellow' , label = 'Cluster 5')

plt.scatter(k.cluster_centers_[:,0], k.cluster_centers_[:,1], s = 100 , c = 'black', label = 'Centroid' )

plt.title(" Clusters of Customers ")
plt.xlabel(" Annual Income ( k$ )")
plt.ylabel(" Spending Score ( 1 - 100 )")
plt.legend()
plt.show()
```
