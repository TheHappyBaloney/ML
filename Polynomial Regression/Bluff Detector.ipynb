```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression


# Data Pre-Processing 
dataset = pd.read_csv('Salary_Data.csv')

x = dataset.iloc[:,1:2].values
y = dataset.iloc[:,2].values

df = pd.DataFrame(dataset)
df
```

```python
# Linear Regression

lin_regs = LinearRegression()
lin_regs.fit(x, y)

predictions = lin_regs.predict(x)

plt.scatter( x , y , color = 'green' )
plt.plot(x,predictions, color = 'red')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.title('Linear Regression')
plt.show()
```

```python
# Polynomial Regression Degree - 2 
from sklearn.preprocessing import PolynomialFeatures

poly_regs = PolynomialFeatures(degree = 2)
x_poly = poly_regs.fit_transform(x)
lin_reg_2 = LinearRegression()
lin_reg_2.fit(x_poly,y)

plt.scatter( x , y , color = 'green' )
plt.plot( x , lin_reg_2.predict( poly_regs.fit_transform ( x ) ), color = 'red')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.title('Polynomial Regression')
plt.show()
```

```python
# Polynomial Regression Degree - 3
from sklearn.preprocessing import PolynomialFeatures

poly_regs = PolynomialFeatures(degree = 3)
x_poly = poly_regs.fit_transform(x)
lin_reg_3 = LinearRegression()
lin_reg_3.fit(x_poly,y)

plt.scatter( x , y , color = 'green' )
plt.plot( x , lin_reg_3.predict( poly_regs.fit_transform ( x ) ), color = 'red')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.title('Polynomial Regression')
plt.show()
```

```python
# Polynomial Regression Degree - 4
from sklearn.preprocessing import PolynomialFeatures

poly_regs = PolynomialFeatures(degree = 4)
x_poly = poly_regs.fit_transform(x)
lin_reg_4 = LinearRegression()
lin_reg_4.fit(x_poly,y)

plt.scatter( x , y , color = 'green' )
plt.plot( x , lin_reg_4.predict( poly_regs.fit_transform ( x ) ), color = 'red')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.title('Polynomial Regression')
plt.show()
```

```python
lin_pred = lin_regs.predict([[6.5]])
print(lin_pred)


poly_pred_4 = lin_reg_4.predict(poly_regs.fit_transform([[6.5]]))
print(poly_pred_4)

```
